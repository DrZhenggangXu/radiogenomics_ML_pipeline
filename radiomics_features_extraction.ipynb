```python
import pandas as pd
import os
basePath = './Segmentation'
folders = os.listdir(basePath)
print(folders)
```

```python
from radiomics import featureextractor
import SimpleITK as sitk
settings = {}
settings['binWidth'] = 25
settings['sigma'] = [3, 5]
settings['resampledPixelSpacing'] = [1,1,1]
settings['interpolator'] = sitk.sitkBSpline
settings['voxelArrayShift'] = 1000
settings['normalize'] = True
settings['normalizeScale'] = 100
extractor = featureextractor.RadiomicsFeatureExtractor(**settings)
```

```python
extractor.enableImageTypeByName('Original')
extractor.enableImageTypeByName('LoG')
extractor.enableImageTypeByName('Wavelet')
extractor.enableImageTypeByName('Square')
extractor.enableImageTypeByName('SquareRoot')
extractor.enableImageTypeByName('Logarithm')
extractor.enableImageTypeByName('Exponential')
extractor.enableImageTypeByName('Gradient')
extractor.enableImageTypeByName('LBP3D')
extractor.enableAllFeatures() 
```

```python
df = pd.DataFrame()
for folder in folders:
    files = os.listdir(os.path.join(basePath,folder))
    print(files)
    for file in files:
        if file.endswith('artery.nii'):
            imageFile_A = os.path.join(basePath,folder,file)
        if file.endswith('artery_roi.nii'):
            maskFile_A = os.path.join(basePath,folder,file)
    print(imageFile_A,maskFile_A)
    featureVector = extractor.execute(imageFile_A,maskFile_A)
    df_new = pd.DataFrame.from_dict(featureVector.values()).T
    df_new.columns = featureVector.keys()
    df_new['sample'] = folder
    df_new = df_new.set_index('sample')
    df = pd.concat([df,df_new])
df.to_excel('./pyradiomics_res/artery_results.xlsx')

df = pd.DataFrame()
for folder in folders:
    files = os.listdir(os.path.join(basePath,folder))
    print(files)
    for file in files:
        if file.endswith('vein.nii'):
            imageFile_V = os.path.join(basePath,folder,file)
        if file.endswith('vein_roi.nii'):
            maskFile_V = os.path.join(basePath,folder,file)
    print(imageFile_V,maskFile_V)
    featureVector = extractor.execute(imageFile_V,maskFile_V)
    df_new = pd.DataFrame.from_dict(featureVector.values()).T
    df_new.columns = featureVector.keys()
    df_new['sample'] = folder
    df_new = df_new.set_index('sample')
    df = pd.concat([df,df_new])
df.to_excel('./pyradiomics_res/vein_results.xlsx')
```
